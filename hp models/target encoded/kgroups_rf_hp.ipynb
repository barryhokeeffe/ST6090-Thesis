#### Set Up

```python
## imports and dependencies

# standard libraries
import pandas as pd
import numpy as np

# models
from sklearn.ensemble import RandomForestRegressor

# folds
from sklearn.model_selection import GroupKFold

# accuracy
from sklearn.metrics import mean_absolute_error

# hyper parametrisation
from sklearn.model_selection import GridSearchCV

# cross validation
from sklearn.model_selection import cross_val_score

# scaler 
from sklearn.preprocessing import StandardScaler  
```

#### Data

```python
# data
df = pd.read_csv("C:/Users/barry/Desktop/tar_enc_dataset.csv", encoding="latin1")
df = df.drop(['Weekly Gross', 'Total Gross', 'Annual Gross', 'Weekly Net', 'Annual Net', 'Ln Annual Net', 'AgeSQ'], axis=1)
```

```python
# target and explanatory variables and groups
IDs = df['Unique ID']
x = df.drop(['Ln Weekly Gross', 'Name', 'Unique ID'], axis=1)
y = df['Ln Weekly Gross']

# as arrays
x = x.values
y = y.values
```

```python
# standard scaler
scaler = StandardScaler()
x = scaler.fit_transform(x)
```

```python
# hyper parametrisation
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt']
}
```

```python
# random forest model
rf = RandomForestRegressor(random_state=42)
```

```python
# grid search w/ group k fold cross validation
group_kfold = GroupKFold(n_splits=5)
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=GroupKFold(n_splits=5))
grid_search.fit(x, y, groups=IDs)
```

```python
# best model
best_rf_model = grid_search.best_estimator_
```

```python
# mean absolute error
y_pred = best_rf_model.predict(x)
mae = mean_absolute_error(y, y_pred)
print("Mean Absolute Error on the whole dataset:", mae)
```

```python
# cross validation 
mae_scores = cross_val_score(best_rf_model, x, y, cv=group_kfold.split(x, y, groups=IDs), scoring='neg_mean_absolute_error')
mae_scores = -mae_scores  # Convert negative scores to positive
print("Cross-Validation Mean Absolute Errors:", mae_scores)
print("Average Mean Absolute Error:", np.mean(mae_scores))
```
