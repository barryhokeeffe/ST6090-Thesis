```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
baseline1 = 0.7193830805118173
baseline4 = 0.3842665269335408
```

```python
dt_scaled = 0.5537165592317733
dt_pca = 0.6918013882126705
```

```python
gb_scaled = 0.36471802167148754
gb_pca = 0.49722253127457056
```

```python
lr_scaled = 0.4271470239143451
lr_pca = 0.4451183173830697
```

```python
mlp_scaled = 0.8712715112659725
mlp_pca = 0.6707908723877025
```

```python
rf_scaled = 0.38258373002758245
rf_pca = 0.5167163353398132
```

```python
knn_scaled = 0.5266126666420213
knn_pca = 0.5477143712730215
```

```python
models = ("Linear Regression","KNN", "Decision Tree", "Random Forest", "Gradient Boosting", "Multi Layer Perceptron")
preprocessing = {
    'Scaled': (lr_scaled, knn_scaled, dt_scaled, rf_scaled, gb_scaled, mlp_scaled),
    'PCA': (lr_pca, knn_pca, dt_pca, rf_pca, gb_pca, mlp_pca),
}

x = np.arange(len(models))  
width = 0.25  
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in preprocessing.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Mean Absolute Error')
ax.set_title('Comparison of Predictive Models [Target Encoding]')
ax.set_xticks(x + width, models, rotation=45)
ax.legend(loc='upper left', ncols=3)

plt.axhline(y=baseline1, color='r', ls='--', label='Baseline')
plt.axhline(y=baseline4, color='r', ls='--')
plt.ylim(0, 1)


plt.legend(loc='upper left')

plt.show()
```
