```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
baseline = 0.3842665269335408
```

```python
dt_scaled = 0.5606384571700185
dt_pca = 0.6662380212655906
```

```python
gb_scaled = 0.3738171199738544
gb_pca = 0.4826548679426255
```

```python
lr_scaled = 0.41843388598172343
lr_pca = 0.4417413596150684
```

```python
mlp_scaled = 0.8268861214293495
mlp_pca = 0.6727904201616102
```

```python
rf_scaled = 0.37848846902236005
rf_pca = 0.5178238444846041
```

```python
knn_scaled = 0.5523121100396121
knn_pca = 0.5523121100396121
```

```python
models = ("Linear Regression","KNN", "Decision Tree", "Random Forest", "Gradient Boosting", "Multi Layer Perceptron")
preprocessing = {
    'Scaled': (lr_scaled, knn_scaled, dt_scaled, rf_scaled, gb_scaled, mlp_scaled),
    'PCA': (lr_pca, knn_pca, dt_pca, rf_pca, gb_pca, mlp_pca),
}

x = np.arange(len(models))  
width = 0.25  
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in preprocessing.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Mean Absolute Error')
ax.set_title('Comparison of Predictive Models [Region + Gen Pos]')
ax.set_xticks(x + width, models, rotation=45)
ax.legend(loc='upper left', ncols=3)

plt.axhline(y=baseline, color='r', ls='--', label='Baseline')

plt.legend(loc='upper left')

plt.show()
```
